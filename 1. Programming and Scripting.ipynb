#### Question 1
Write a program that asks the user to input any positive integer and outputs the sum of all numbers between one and that number.

```python
# Asks the user to input a positive integer number 

i = int(input("Please enter a positive integer: "))

# This will prevent the user from entering anything other than a positive integer.
# The message will print when the user types a negative integer.
if i <= 0: 
  print("Unfortunately this is not a positive integer. Please, try again.")

total = 0 

while i > 0:  # a while loop 
    total = total + i
    i = i - 1
    
# While 'i' is greater than 0 add the total and the current value of "i" and overwrite the current total. 

# i = i - 1: If the user inputs the positive integer of 10, the program subtracts one from the current value and so on, down to 0.
 
print(total)
```

This code will request the user to input a positive number. Once the number is entered it will get the total sum between 1 and the user input. I have used a function to accept the user input, used **while** and **if** statements to retrieve the required output.

If a negative number is entered the user will be informed that she/he can not use a negative number and if she/he would like to try again. I have used a function that will accept the number entered by the user.

#### Question 7
Write a program that that takes a positive floating point number as input and outputs an approximation of its square root.


```python
# The number to calculate the square root of 14.5
rootof = 14.5 
# The initial estimate for the square root
estimate = 1.0


while abs((estimate * estimate) - rootof) > 0.1:
    estimate -= (( estimate * estimate) - rootof) / (2 * estimate)

print(f"The square root of {rootof} is approx. {estimate}.")
```
